```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('sin(x) and sin(-x)')
ax1.plot(x, y, "tab:green")
ax2.plot(x, -y, "tab:orange")
```

```python
y_ = x ** np.sin(x ** 2)
```

```python
fig_, axs_ = plt.subplots(2)
fig_.suptitle('xsin(x) and -xsin(-x)')
axs_[0].plot(x, y_, 'tab:purple')
axs_[1].plot(x, -y_, 'tab:pink')
```

```python
import panel as pn
pn.extension()
```

```python
tabs = pn.Tabs(('Scatter 1', fig),("Scatter 2", fig_))
```
